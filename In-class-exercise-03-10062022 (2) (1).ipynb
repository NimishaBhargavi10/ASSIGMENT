Question 2 (20 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction. 

```python
# You code here (Please add comments in the code):
#Converting entire text into lower case characters

from six.moves import urllib
import pandas as pd

t1 = []
url = "https://raw.githubusercontent.com/NimishaBhargavi10/ASSIGMENT/main/SampleText.txt"
tf = urllib.request.urlopen(url)  #This function is used to decode the url and get the data from it it is a predefined function

for i in tf:
    dec_data = i.decode("utf-8").replace('\r\n', '')    #this statement is used to decode the data present in the url and convert it into standard format
    if dec_data: 
        t1.append(dec_data)       #appending the decoded data to a list
df = pd.DataFrame (t1, columns = ['Text'])   #converting the list to data frame using pamdas

#1Lowercase
df["Text"]=df["Text"].str.lower()
df







```

```python
#2removing all punctuations and unnecessary symbols from text
import string
for char in string.punctuation:
    df.Text=df.Text.replace(char,'')
print("1) Remove noise, such as special characters and punctuations \n", df.Text.head(), end="\n\n")
df
```

```python
#3removing numerical characters from the text 
df.Text = df.Text.str.replace('\d+','')
print(" Remove numbers \n", df.Text.head(), end="\n\n")
df
```

```python
#4 Count number of special chracters, words

df['Number of Words'] = df['Text'].apply(lambda y: len(str(y).split(" ")))

df['Number of Characters'] = df['Text'].str.len()


def spchar(x):
    n = 0
    for i in range(len(x)):
        if not((x[i].isalpha()) and (x[i].isdigit())):
            n = n + 1
    return n

df['Num of special char'] = df['Text'].apply(lambda y: spchar(y))
df
```

```python
#5find number of lowercase
df['Num of lower case words'] = df['Text'].apply(lambda x: len([x for x in x.split() if x.islower()]))
df
```
