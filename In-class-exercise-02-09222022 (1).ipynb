Question 2 (10 points): Write python code to collect 1000 data samples you discussed above.

```python
# You code here (Please add comments in the code):
import requests
from bs4 import BeautifulSoup
import pandas as pd
review = [] # List to store Review headings
review_description =[] #List to store reviews
for number in range(102):
    link = "https://www.flipkart.com/fossil-briggs-analog-watch-men/p/itmf3zhfhfmfqmrk?pid=WATE4U6GNBTMHYTQ&ssid=4oe0ol5ijk0000001664166049431" + str(number) # Generating link dynamically
    page = requests.get(link) # Accessing the webpage
    soup = BeautifulSoup(page.text, 'html.parser')
    review = soup.find_all(class_='_2-N8zT') # Getting the Review Heading by using the class name
    review_description = soup.find_all(class_='_3LWZlK _1BLPMq') # Getting the full reviews by using the class name
    for ele, sub_ele in zip(review, review_description) : # Iterating through the list
        review.append(ele.text) #Appending to empty list
        review_description.append(sub_ele.text)
apple_review_df = pd.DataFrame(list(zip(review,review_description )), columns =['product review', 'description'])  # Dataframe creating
print("Length of data frame is {0}".format(len(apple_review_df)))
apple_review_df


```
